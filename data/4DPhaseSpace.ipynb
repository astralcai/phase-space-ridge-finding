```python
import numpy as np
import matplotlib.pyplot as plt
import corner as corner
```

```python
data = np.loadtxt('P_DeltaRc',unpack=True,skiprows=1)
R = np.sqrt(data[1,:]**2 + data[2,:]**2)
z = data[3,:]
vz = data[6,:]
vR = (data[1,:]*data[4,:] + data[2,:]*data[5,:])/R
vp = (data[1,:]*data[5,:] - data[2,:]*data[4,:])/R
Lz = vp*R
samples = np.stack((R,z,vR,vz)).transpose()
corner.corner(samples)
```

```python
data = np.loadtxt('P_DeltaRc.0500.ascii',unpack=True,skiprows=1)
R = np.sqrt(data[1,:]**2 + data[2,:]**2)
z = data[3,:]
vz = data[6,:]
vR = (data[1,:]*data[4,:] + data[2,:]*data[5,:])/R
vp = (data[1,:]*data[5,:] - data[2,:]*data[4,:])/R
Lz = vp*R
samples = np.stack((R,z,vR,vz)).transpose()
corner.corner(samples,labels=[r"$R$",r"$z$",r"$v_R$",r"$v_z$"],label_kwargs={"fontsize": 18},plot_contours=None,bins=100)
```
